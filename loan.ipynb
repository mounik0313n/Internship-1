```python
# Columns to visualize (excluding non-numeric columns)
columns_to_visualize = [
    'Age', 'Income (USD)', 'Income Stability', 
    'Current Loan Expenses (USD)', 'Dependents', 
    'Credit Score', 'Property Age', 'Loan Sanction Amount (USD)'
]

# Function to detect outliers using Z-score
def detect_outliers_zscore(data, columns):
    outliers = pd.DataFrame()
    for col in columns:
        if np.issubdtype(data[col].dtype, np.number):  # Check if column is numeric
            z_scores = np.abs((data[col] - data[col].mean()) / data[col].std())
            outliers[col] = data[col][z_scores > 3]  # Adjust the threshold as needed
    return outliers

# Detecting outliers
outliers_zscore = detect_outliers_zscore(df, columns_to_visualize)

print("Outliers detected using Z-score method:")
print(outliers_zscore)

# Visualizing outliers with box plots
plt.figure(figsize=(14, 10))
for i, col in enumerate(columns_to_visualize, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)

plt.tight_layout()
plt.show()
```

```python
# Columns to check skewness (excluding non-numeric columns)
numeric_columns = df.select_dtypes(include=['number']).columns

# Determine the number of rows and columns for subplots
num_columns = len(numeric_columns)
num_rows = (num_columns // 4) + (num_columns % 4 != 0)  # Ensure enough rows for all columns

plt.figure(figsize=(20, 5 * num_rows))
for i, col in enumerate(numeric_columns, 1):
    plt.subplot(num_rows, 4, i)
    sns.histplot(df[col], kde=True, color='skyblue', stat='density')
    plt.title(f'Skewness: {df[col].skew():.2f}\n{col}')
    plt.xlabel('')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Example list of numerical columns to check for outliers
columns_to_check = [
    'Age', 'Loan Amount Request (USD)', 'Current Loan Expenses (USD)',
    'Dependents', 'Credit Score', 'Property Age', 'Property Price',
    'Loan Sanction Amount (USD)'
]

# Define function to detect outliers using IQR method
def detect_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data < lower_bound) | (data > upper_bound)]
    return outliers

# Create a figure to plot multiple subplots
plt.figure(figsize=(16, 12))

# Loop through each column and plot boxplot for outliers
for i, col in enumerate(columns_to_check, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(x=df[col])
    plt.title(f'Box Plot of {col}')
    plt.xlabel(col)

plt.tight_layout()
plt.show()

# Detect outliers for each column using IQR method
outliers = {}
for col in columns_to_check:
    outliers[col] = detect_outliers_iqr(df[col])

# Print outliers detected (optional)
print("Outliers detected using IQR method:")
for col, out in outliers.items():
    if not out.empty:
        print(f"{col}:")
        print(out)
        print()

# Optionally, remove outliers from the DataFrame
# for col, out in outliers.items():
#     df = df[~df[col].isin(out)]

# Print DataFrame after removing outliers (optional)
# print("DataFrame after removing outliers:")
# print(df)
```
